```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = school_data_complete["school_name"].nunique()
#Calculate the total number of students
total_students = school_data_complete["Student ID"].nunique()
#Calculate the total budget
total_budget = school_data["budget"].sum()
#Calculate the average math score and reading score
average_math_score = school_data_complete["maths_score"].mean() 
average_reading_score = school_data_complete["reading_score"].mean()

#Calculate the percentage of students with a passing math score (50 or greater)

limit = 50
mscore = 'maths_score'
# Select column 'C' from the dataframe
column = school_data_complete[mscore]
# Get count of values greater than 50 in the column 'math_score'
count = column[column > limit].count()
math_passing = count / total_students *100


#Calculate the percentage of students with a passing reading score (50 or greater)

limit = 50
rscore = 'reading_score'
# Select column 'C' from the dataframe
column = school_data_complete[rscore]
# Get count of values greater than 50 in the column 'math_score'
count = column[column > limit].count()
read_passing = count / total_students *100


```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
# df.loc[(df['team'] == 'A') & (df['conference'] == 'East'), 'points'].sum()
df = school_data_complete
overall_passing_rate = df.loc[(df['maths_score'] > limit) & (df['reading_score'] > limit), "Student ID"].count()/ total_students *100
overall_passing_rate
```

## School Summary

```python
# Store all calculations into dataframe
district_summary = pd.DataFrame({"Total Schools":[total_schools],
                                 "Total Students":[total_students],
                                 "Total Budget":[total_budget],
                                 "Average Math Score":[average_math_score],
                                 "Average Reading Score":[average_reading_score],
                                 "% Passing Math":[math_passing],
                                 "% Passing Reading":[read_passing],
                                 "% Overall Passing Rate":[overall_passing_rate]})
district_summary
```

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school
# School Name
grouped_school_df = school_data_complete.groupby(["school_name"])
# School Type
schoolType = grouped_school_df["type"].first()
# Total Students
totalStudents = grouped_school_df["Student ID"].count()
# Total School Budget
totalSchoolBudget = grouped_school_df["budget"].first()
# Per Student Budget
perStudentBudget = totalSchoolBudget / totalStudents
# Average Math Score
averageMathScore = grouped_school_df["maths_score"].mean()
# Average Reading Score
averageReadingScore = grouped_school_df["reading_score"].mean()
# % Passing Math
passingMath = school_data_complete[school_data_complete["maths_score"] >= 70].groupby(["school_name"])["maths_score"].count()
percentagePassingMath = passingMath / totalStudents * 100
# % Passing Reading
passingRead = school_data_complete[school_data_complete["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
percentagePassingReading = passingRead / totalStudents * 100
# Overall Passing Rate (Average of the above two)
percentageOverallPassingRate = (percentagePassingMath + percentagePassingReading) / 2


# Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type": schoolType,
      "Total Students": totalStudents,
      "Total School Budget": totalSchoolBudget,
      "Per Student Budget": perStudentBudget,
      "Average Math Score": averageMathScore,
      "Average Reading Score": averageReadingScore,
      "% Passing Math": percentagePassingMath,
      "% Passing Reading": percentagePassingReading,
      "% Overall Passing Rate": percentageOverallPassingRate})
school_summary_df

```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.
#year9_ds = school_data_complete.loc[school_data_complete["year"] == 9].groupby(["school_name"])["maths_score"].mean()

year_ds = school_data_complete.loc[school_data_complete["year"] > 8].groupby(["school_name","year"])["maths_score"].mean()

#Create a pandas series for each year. Hint: use a conditional statement.
#Group each series by school
year9_ds = school_data_complete.loc[school_data_complete["year"] == 9].groupby(["school_name"])["maths_score"].mean()
year10_ds = school_data_complete.loc[school_data_complete["year"] == 10].groupby(["school_name"])["maths_score"].mean()
year11_ds = school_data_complete.loc[school_data_complete["year"] == 11].groupby(["school_name"])["maths_score"].mean()
year12_ds = school_data_complete.loc[school_data_complete["year"] == 12].groupby(["school_name"])["maths_score"].mean()

#Combine the series into a dataframe
year_summary_df = pd.DataFrame({9: year9_ds,
      10: year10_ds,
      11: year11_ds,
      12: year12_ds})

year_summary_df[[9, 10, 11, 12]]


#Optional: give the displayed data cleaner formatting
```
